```python
import numpy as np
import pandas as pd
```

```python
# Load the data
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df
```

```python
# initial inspection
df.shape
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
# Checking missing values
df.isnull().sum()
```

```python
df.isnull().mean() * 100 #  % missing per column
```

```python
# Handling Missing values
df["Cabin"]
```

```python
# Drop colums with too many missing values
df.drop(columns=['Cabin'], inplace = True)
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Filling numeric column with median
df["Age"].fillna(df["Age"].median(), inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Remove Duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Fix datatypes
df["Survived"] = df["Survived"].astype('bool')
df["Pclass"] = df["Pclass"].astype('category')
```

```python
df.dtypes
```

```python
df["Sex"].head()
```

```python
df["Sex"] = df["Sex"].str.lower().str.strip()
df["Sex"].head()
```

```python
df["Fare"].head()
```

```python
# Handlinhg outliers
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

```

```python
df = df[(df['Fare'] >= Q1 - 1.5*IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]
df
```

```python
# Rename for clarity
df.rename(columns = {"Pclass": "PassengerClass"}, inplace = True)
df
```

```python
# Final Check
df.isnull().sum()
```
